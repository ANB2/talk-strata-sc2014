```python
from __future__ import print_function
from IPython.utils.traitlets import Float, Unicode, Int, link
from IPython.html import widgets
from IPython.html.widgets import interact, interactive, fixed
from IPython.display import display
```

## The Leap Widget implementation

Here is the Python side of the Widget:

```python
class LeapWidget(widgets.DOMWidget):
    _view_name = Unicode('LeapWidgetView', sync=True)
    nfingers = Int(sync=True)
    xx = Float(sync=True) # Normalized to [0,1]
    yy = Float(sync=True) # Normalized to [0,1]
    zz = Float(sync=True) # Normalized to [0,1]
    x = Float()           # Scaled to [xmin,xmax]
    xmin = Float(0.0)
    xmax = Float(1.0)
    y = Float()           # Scaled to [ymin,ymax]
    ymin = Float(0.0)
    ymax = Float(1.0)
    z = Float()           # Scaled to [zmin,zmax]
    zmin = Float(0.0)
    zmax = Float(1.0)
    
    def _xx_changed(self, name, new):
        d = self.xmax-self.xmin
        self.x = new*d+self.ymin
    
    def _yy_changed(self, name, new):
        d = self.ymax-self.ymin
        self.y = new*d+self.ymin
    
    def _zz_changed(self, name, new):
        d = self.zmax-self.zmin
        self.z = new*d+self.zmin
```

Here is the JavaScript side of the Widget:

```python
%%javascript

$.getScript("//js.leapmotion.com/0.2.0/leap.min.js");

require(["notebook/js/widgets/widget"], function(WidgetManager){
    
    var LeapWidgetView = IPython.DOMWidgetView.extend({
        
        initialize: function (options) {
            this.controller = new Leap.Controller();
            this.controller.on('ready', $.proxy(this.handle_ready, this));
            this.controller.connect();
            LeapWidgetView.__super__.initialize.apply(this, arguments);
        },
        
        render: function () {
            this.$readout = $('<span/>').appendTo(this.$el);
        },
                
        handle_ready: function () {
            var that = this;
            this.timer = setInterval(function () {
                that.handle_frame(that.controller.frame());
            }, 100);
        },
        
        remove: function () {
            clearInterval(this.timer);
            LeapWidgetView.__super__.remove.apply(this, arguments);
        },
        
        handle_frame: function (frame) {
            var nfingers = frame.fingers.length;
            var updates = {};
            updates['nfingers'] = nfingers;
            var hand = frame.hands[0];
            if (hand) {
                var pos = this.normalize_position(frame, hand.palmPosition);
                var x = pos[0];
                var y = pos[1];
                var z = pos[2];
                updates['xx'] = x;
                updates['yy'] = y;
                updates['zz'] = z;                
                this.$readout.text(x.toFixed(3) + ', ' + y.toFixed(3) + ', ' + z.toFixed(3));
            }
            this.model.set(updates);
            this.touch();
        },
        
        clip: function (value) {
            if (value < 0.0) {return 0.0;}
            else if (value > 1.0) {return 1.0;}
            else {return value;}
        },
        
        normalize_position: function (frame, leap_pos) {
            var ibox = frame.interactionBox;
            var x = (leap_pos[0]/ibox.size[0]+1)/1.5;
            var y = 1.3*leap_pos[1]/ibox.size[1]-0.5;
            var z = 2.2*(leap_pos[2]/ibox.size[2]+1)/2-0.9;
            x = this.clip(x);
            y = this.clip(y);
            z = this.clip(z);
            return [x, y, z];
        }
    });
    
    // Register the DatePickerView with the widget manager.
    WidgetManager.register_widget_view('LeapWidgetView', LeapWidgetView);
});
```

## Use the Widget to explore a 4d dataset

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
def make_scatter(data, col1, col2):
    X = data.data
    y = data.target
    plt.scatter(X[:,col1], X[:,col2], c=y, s=40, alpha=0.4)
    plt.show()
```

```python
make_scatter(data, 1, 0)
```

```python
l = LeapWidget()
l.xmin = 0.0
l.xmax = 3.0
l.ymin = 0.0
l.ymax = 3.0
```

```python
w = interactive(make_scatter, data=fixed(data), col1=(0,3), col2=(0,3))
```

```python
s0 = w.children[0]
s1 = w.children[1]
link((s0,'value'), (l,'x'))
link((s1,'value'), (l,'y'))
display(w)
display(l)
```

```python
w.close()
l.close()
```

## Link the Widget to a matplotlib 3d plot

```python
%matplotlib inline
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
# plt.rcParams['figure.figsize'] = (10, 8)
```

```python
n_angles = 36
n_radii = 8
radii = np.linspace(0.125, 1.0, n_radii)
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)
angles = np.repeat(angles[...,np.newaxis], n_radii, axis=1)
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())
z = np.sin(-x*y)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_trisurf(x, y, z, cmap=cm.jet, linewidth=0.2);
ax.set_axis_off()
```

```python
l = LeapWidget()
l.xmin = 0.0
l.xmax = 360.0
l.ymin = -90.0
l.ymax = 90.0
```

```python
def view(azim, elev):
    ax.view_init(90-elev, 360-azim)
    display(ax.figure)
```

```python
w = interactive(view, azim=(0.0, 360.0, 18.0), elev=(-90.0, 90.0, 9.0))
```

```python
s0 = w.children[0]
s1 = w.children[1]
link((s0,'value'), (l,'x'))
link((s1,'value'), (l,'y'))
display(w)
display(l)
```

```python
w.close()
l.close()
```
